#### 3.1-Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)

Hypothesis

Null- No difference in median value between homes bounded by Charles river and those that are not

Alternate - Significant difference in median value between homes bounded by Charles river and those that are not

```python
boston_df.loc[(boston_df['CHAS'] == 0), 'CHAS_T'] = 'FAR'
boston_df.loc[(boston_df['CHAS'] == 1), 'CHAS_T'] = 'NEAR'
boston_df.head(5)
```

```python
scipy.stats.ttest_ind(boston_df[boston_df['CHAS_T'] == 'FAR']['MEDV'], 
                      boston_df[boston_df['CHAS_T'] == 'NEAR']['MEDV'], equal_var = True)
```

As p-value<0.05, null hypothesis is resject, so there is a significant difference in the median value between homes bounded by the Charles river and homes not bounded by the Charles River

#### 3.2- Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)

Hypothesis

Null- No difference in  Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940

Alternate - There is difference in  Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940

```python
from statsmodels.formula.api import ols
xy = ols('MEDV ~ AGE', data = boston_df).fit()
table = sm.stats.anova_lm(xy)
print(table)
```

As p-value<0.05, null is rejected.There is statistical difference in Median values of houses (MEDV) for each proportion of owner occpied units built prior to 1940

#### 3.3- Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town? (Pearson Correlation)

Hyposthesis:

Null: There is no correlation between NOx levels and proportion of non retail businesses in town

Alternate: There is correlation between NOx levels and proportion of non retail businesses in town

```python
scipy.stats.pearsonr(boston_df['NOX'], boston_df['INDUS'])
```

```python
xy = sns.scatterplot(y = 'NOX', x = 'INDUS', data = boston_df)
xy.set_title('Nitric oxide concentration per non-retail business acres per town')
```

Given a p-value<0.05, a Pearson of 0.763 it is possible to reject the null hypothesis and conclude there is a correlation between NOx levels and non-retail business acres per town. Also supported by looking at the scatter plot

#### 3.3- What is the impact of an additional weighted distance to the five Boston employment centres on the median value of owner occupied homes? (Regression analysis)

```python
x = boston_df['DIS']
y = boston_df['MEDV']

x = sm.add_constant(x)

xy = sm.OLS(y, x).fit()
prediction = xy.predict(x)

xy.summary()
```
